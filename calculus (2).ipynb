```python
# A program to do linear regression

# Import numpy to aid with some maths
import numpy as np

# Two arrays filled with random numbers
X = np.array([1, 2.5, 3, 3.25, 6, 8, 9.4])
y = np.array([0, 5, 7, 6.5, 9.5, 23.5, 18.7])

# 4 functions for doing linear regression manually
# Model function that returns m * x
def model(x, m):
    return m * x + 0 # mx + b [b = 0]

# Error function
def error_function(m):

    # Define the numerator of the error function
    predictions = model(X, m)

    n = len(y)
    return 1/n * np.sum((y - predictions) ** 2)

# Derivative function
def derivative(m):

    # J' = 1/n * sum -2 * (x * (yi - yp))

    summation = 0
    n = len(X)

    for i in range(n):
        x = X[i]
        y_true = y[i]

        # Calculate inner expression
        error = y_true - model(x, m)
        summation += -2 * x * error

    return summation / n

# Update function
def update_step(m, learning_rate=0.01):

    derivative_term = derivative(m)
    updated_amount = learning_rate * derivative_term

    # Change m
    return m - updated_amount

# The main program
m = -6
n_epochs = 5 # This is just the number of update steps you take

# Iterate for n epochs
for epoch in range(n_epochs):
    # Update and print m
    m = update_step(m)
    print(f'Epoch {epoch + 1}: m = {round(m,2)} and loss = {round(error_function(m), 2)}')

# Print Final Results
print("-----------------")
print("FINAL RESULTS")
print("-----------------")

print(f'm = {round(m,2)} and loss = {round(error_function(m), 2)}')

print("-----------------")
```
